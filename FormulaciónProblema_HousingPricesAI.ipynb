# Formulación del Problema

## 1. Definición clara del problema.

El problema consiste en contar parametros establecidos para una posible vivienda, parametros tales como, tamaño de la vivienda (N metros cuadrados), cantidad de cuartos, ubicación, y demás parametros importantes para aplicar los diferentes modelos, se toma un DataSet con información de viviendas con diferentes caracteristicas y lograr generar mediante procesamiento de datos y algoritmos matematicos, un valor estimado lo más cercano al real

## 2. Tipo de aprendizaje (supervisado o no supervisado).

El tipo de aprendizaje es supervisado, esto ya que se trabaja con datos historicos ya etiquetados 

## 3. Variables de entrada y salida.

Las variable de entrada otorgadas por el usuario son las siguientes:
- Año de fabricación
- Tamañano de la vivienda (N metros cuadrados)
- Número de pisos
- Cantidad de habitaciones
- Cantidad de baños
- Ubiciación 

La salida esperada es la siguiente:
- El precio más cercano posible al real de la vivienda

Se debe tomar en cuenta que aunque se usen dos modelos, la Regresión Lineal y el Random Forest, al final, con medidas de aceptación se tomará como más apto uno de ambos modelos, o una combinación de amboa

## 4. Justificación técnica del enfoque elegido.
Se elige un enfoque de regresión supervisada porque el objetivo es predecir un valor numérico continuo (el precio), no clasificar en categorías. Además se implementan dos modelos complementarios:

#### Regresión Lineal: 
Modelo base interpretable que asume una relación lineal entre las variables de entrada y el precio. Permite entender el peso de cada variable y sirve como línea base de comparación.
#### Random Forest Regressor: 
Modelo de ensemble basado en múltiples árboles de decisión que captura relaciones no lineales y es más robusto ante valores atípicos. Se espera que supere a la regresión lineal cuando las relaciones entre variables no sean estrictamente lineales.
