# Formulación del Problema

### 1. Definición clara del problema.

El problema consiste en tener parametros establecidos para una posible vivienda, parametros como tamaño de la vivienda (x metros cuadrados), cantidad de cuartos, ubicación, entro otros, con estos parametros, tomar un DataSet con información de viviendas con diferentes caracteristicas y lograr generar mediante procesamiento de datos y algoritmos matematicos, devolver un valor estimado lo más cercano al real

Las variable de entrada otorgadas por el usuario son las siguientes:
- Tamañano de la vivienda en metros cuadrados
- Cantidad de habitaciones
- Ubiciación 

La salida esperada es la siguiente:
- El precio más cercano posible al real de la vivienda
